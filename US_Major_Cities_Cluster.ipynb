#### This function will then pull each text from each box and place it in a list as the function runs through each column

```python
#Rank
A= []
#City
B= []
#Population Estimate
C= []
#2010 Census
D= []
#2000 Census
E= []
#1990 Census
F= []

#For loop function that for every row in the rightable row <tr>
# it will append each cell into the empty cell. 

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==6:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
        D.append(cells[3].find(text=True))
        E.append(cells[4].find(text=True))
        F.append(cells[5].find(text=True))
```

#### This function will extract the categories from each row when used, and let you know if there none if none listed

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

#### This function will now run through all Major Cities and pull venues in its immediate downtown area

```python
def getNearbyVenues(names, latitudes, longitudes, radius=2000):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['City', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
#one hot encoding
TopTen_onehot = pd.get_dummies(TopTen_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
TopTen_onehot['City'] = TopTen_venues['City'] 

# move neighborhood column to the first column
fixed_columns = [TopTen_onehot.columns[-1]] + list(TopTen_onehot.columns[:-1])
TopTen_onehot = TopTen_onehot[fixed_columns]

TopTen_onehot.head()
```

```python
TopTen_onehot.shape
```

```python
TopTen_grouped = TopTen_onehot.groupby('City').mean().reset_index()
TopTen_grouped
```

```python
TopTen_grouped.shape
```

```python
num_top_venues = 5

for hood in TopTen_grouped['City']:
    print("----"+hood+"----")
    temp = TopTen_grouped[TopTen_grouped['City'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['City']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
TopTen_venues_sorted = pd.DataFrame(columns=columns)
TopTen_venues_sorted['City'] = TopTen_grouped['City']

for ind in np.arange(TopTen_grouped.shape[0]):
    TopTen_venues_sorted.iloc[ind, 1:] = return_most_common_venues(TopTen_grouped.iloc[ind, :], num_top_venues)

TopTen_venues_sorted
```

```python
# set number of clusters
kclusters = 5

TopTen_grouped_clustering = TopTen_grouped.drop('City', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(TopTen_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels
TopTen_venues_sorted.insert(0, 'Cluster Label', kmeans.labels_)

TopTen_merged = TopTen

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
TopTen_merged = TopTen_merged.join(TopTen_venues_sorted.set_index('City'), on='City')

TopTen_merged # check the last columns!
```

```python
TopTen_merged = TopTen_merged[TopTen_merged['Cluster Label'].notnull()]
```

```python
TopTen_merged.dtypes
```

```python
TopTen_merged['Cluster Label'] = TopTen_merged['Cluster Label'].apply(np.int32)
```

```python
# create map
map_clusters = folium.Map(location=[40, -100], zoom_start=4)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(TopTen_merged['Latitude'], TopTen_merged['Longitude'], TopTen_merged['City'], TopTen_merged['Cluster Label']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=8,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=1).add_to(map_clusters)
       
map_clusters
```
